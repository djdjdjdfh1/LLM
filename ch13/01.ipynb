seq2seq
```
영문장을 한국어로 번역, 질의 --> 답변생성, 문장요약등.
why?
    입력과 출력의 길이가 다를수 있는 문제(번역, 요약, 챗봇)
    고정크기의 피쳐벡터는 순서정보와 문맥관계를 충분히 반영 못함
    단순 분류모델은 시퀀스 간 종속적 생성에 부적합
```
동작원리
```
    1.Encoder: 입력토큰을 순차적으로 처리 -> 마지막/전체 Hidden State 집약
    2.Context(요약표현): 인코더의 정보를 압축 
    3.Decoder: Context + 이전에 생성한 토큰을 이용해서 다음 토큰을 반복생성
    4.종료: 특별한 EOS 토큰이 나올때까지
```
실제사용예시
```
번역, 대화시스템, 문서요약, 코드자동생성
```
