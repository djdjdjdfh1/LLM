```python
# RAG   데이터증강, llm이 학습하지 않은 기업내 내부문서 등과 같은 자료에서 데이터를 추출해서 llm전달
    # llama_index : 사용하기 쉬움
# ChromaDB : 문장들의 유사성 기반으로 데이터 검색 - 벡터화된 데이터를 검색
# openai - finetuning
# cache : 잦은 llm호출을 방지하고자 정확하거나 의미가 유사한 질문은 기존 답을 재 활용
```

```python
# RAG 기초
# LLM 캐시
# 가드레일(데이터 검증)
    # 입력검증
    # 출력검증
    # 프롬프트 인젝션 방어

# 입력 ->[가드레일 검증]       ---> LLM  -->[가드레일 검증] ---> 출력
       # 안전하지 않으면 차단               # 안전하지 않으면 차단  

#  https://platform.openai.com/docs/guides/moderation
```

```python
from openai import OpenAI
client = OpenAI()
def check_moderation(text):
    response = client.moderations.create(
        model="omni-moderation-latest",
        input=text,
    )
    return response

result = check_moderation("나는 배고파 죽을거 같다")
```

```python
result.results[0].flagged
```

```python
for category, detected in result.results[0].categories.__dict__.items():
    if detected:
        print(category)
```
